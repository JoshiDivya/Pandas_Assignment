## Cleaning

### Drop all the missing rows which contain any column with missing data

```python
df.dropna()
```

```python
df1 = df['duration'].str.split(' mins.', expand=True)
df['duration']=(df1[0].astype(int))*60
df.head()
```

```python
s=pd.Series(df['title'])
df['subtitle']=s.map(lambda a: a[:-6])
df['title']=df['subtitle']
df.head()
```

```python
df=df.drop(columns='subtitle')
df.head()
```
